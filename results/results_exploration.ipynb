```python
import os
import pickle
import pandas as pd

# Define the directory containing your pickle files
# Replace '.' with your folder path if needed (e.g., 'path/to/pickles')
directory_path = "pickles"

# List to store data before converting to a DataFrame
data_list = []
merged_df = pd.DataFrame()
# Loop through all files in the directory
for filename in os.listdir(directory_path):
    # Check if the file has a .pickle or .pkl extension
    if filename.endswith(".pickle"):
        file_path = os.path.join(directory_path, filename)

        try:
            # Open and load the pickle file
            with open(file_path, "rb") as f:
                data = pd.DataFrame(pickle.load(f))
            merged_df = pd.concat([merged_df,data], ignore_index=True)
            # Ensure the loaded data is a dictionary (or supports len())
            if isinstance(data, dict):
                file_length = len(data)
            # Append the name and length to our list
            data_list.append({"File Name": filename, "Length": file_length})

        except Exception as e:
            print(f"Error reading {filename}: {e}")

# Convert the list of dictionaries into a pandas DataFrame
df = pd.DataFrame(data_list)

# Display the DataFrame
print(df)
```

```python
import re

def strip_thinking(text):
    if not isinstance(text, str):
        return text

    # Remove everything before </think>
    text = re.sub(r"(?is)^.*?</think>\s*", "", text).strip()

    # Remove wrapping quotes if present
    text = text.strip("'\"")

    return text


def clean_reasoning_responses(df):
    target_models = [
        "Qwen3.5-122B-A10B",
        "INTELLECT-3",
    ]

    mask = df["model"].isin(target_models)

    df.loc[mask, "response"] = (
        df.loc[mask, "response"]
        .apply(strip_thinking)
    )

    return df
```

```python
def validate_prompt1_responses(df):
    mask = df["prompt_nr"] == 1
    
    allowed = {"CLONE", "NOT CLONE", "- CLONE", "- NOT CLONE", "**CLONE**", "**NOT CLONE**"}
    invalid = df.loc[mask, "response"].isin(allowed) == False
    
    if invalid.any():
        bad_rows = df.loc[mask].loc[invalid, ["prompt_nr", "response"]]
        print(bad_rows[::-1].iloc[0]["response"])
        raise ValueError(
            f"Invalid responses found (only 'CLONE' or 'NOT CLONE' allowed):\n{bad_rows}"
        )
    
    return True
def check_prompt1_contains_clone(df):
    mask = df["prompt_nr"] == 1
    
    bad_mask = ~df.loc[mask, "response"].astype(str).str.contains("CLONE", na=False)
    
    if bad_mask.any():
        bad_rows = df.loc[mask].loc[bad_mask, ["prompt_nr","model", "response"]]
        raise ValueError(
            f"Found prompt_nr == 1 rows without 'CLONE' in response:\n{bad_rows}"
        )
    
    return True
```

```python
def add_model_response_prompt(df, prompt_nr, clone_string):
    mask = df["prompt_nr"] == prompt_nr

    def classify(text):
        if not isinstance(text, str):
            return "Model nie odpowiedział"

        # normalize spacing (handles weird variants like multiple spaces, newlines, etc.)
        clean = re.sub(r"\s+", " ", text)

        # find all CLONE occurrences
        clone_positions = [m.start() for m in re.finditer(re.escape(clone_string), clean)]

        if not clone_positions:
            return "Model nie odpowiedział"

        last_clone_pos = clone_positions[-1]

        # check if immediately before it is NOT CLONE
        before = clean[:last_clone_pos].strip()

        if before.endswith("NOT"):
            return 0

        return 1

    df.loc[mask, "model_response"] = df.loc[mask, "response"].apply(classify)

    return df
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patheffects as path_effects

df = direct_responses_df.copy()

# =========================================================
# 1. PREPROCESS
# =========================================================

def parse_pred(x):
    if pd.isna(x):
        return "NR"
    x = str(x).strip()

    if "Model nie odpowiedział" in x:
        return "NR"

    if x in ["1", "1.0", "True", "TRUE", "clone", "CLONE", "YES", "yes"]:
        return 1
    if x in ["0", "0.0", "False", "FALSE", "not clone", "NOT CLONE", "NO", "no"]:
        return 0

    try:
        return int(float(x))
    except:
        return "NR"


df["y_pred"] = df["model_response"].apply(parse_pred)
df["y_true"] = df["is_clone"].astype(int)


# =========================================================
# 2. METRYKI PER MODEL (F1 / PREC / REC / ACC / SPEC)
# =========================================================

def metrics(y_true, y_pred):
    y_true = np.array(y_true)
    y_pred = np.array(y_pred)

    tp = np.sum((y_true == 1) & (y_pred == 1))
    tn = np.sum((y_true == 0) & (y_pred == 0))
    fp = np.sum((y_true == 0) & (y_pred == 1))
    fn = np.sum((y_true == 1) & (y_pred == 0))

    precision = tp / (tp + fp) if (tp + fp) else 0
    recall = tp / (tp + fn) if (tp + fn) else 0
    f1 = (2 * precision * recall / (precision + recall)) if (precision + recall) else 0
    acc = (tp + tn) / len(y_true)
    specificity = tn / (tn + fp) if (tn + fp) else 0

    return precision, recall, f1, acc, specificity


results = []

for m in df["model"].unique():
    sub = df[df["model"] == m]

    # ignorujemy NR w metrykach klasycznych
    sub_clean = sub[sub["y_pred"] != "NR"]

    if len(sub_clean) == 0:
        continue

    p, r, f1, acc, spec = metrics(sub_clean["y_true"], sub_clean["y_pred"])

    results.append({
        "model": m,
        "precision": p,
        "recall": r,
        "f1": f1,
        "accuracy": acc,
        "specificity": spec,
        "nr_rate": (sub["y_pred"] == "NR").mean(),
        "n": len(sub)
    })

metrics_df = pd.DataFrame(results).sort_values("f1", ascending=False)

print(metrics_df)


# =========================================================
# 3. RANKING MODELI (F1)
# =========================================================

plt.figure()
plt.bar(metrics_df["model"], metrics_df["f1"])
plt.title("Ranking modeli (F1-score)")
plt.xlabel("Model")
plt.ylabel("F1")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()


# =========================================================
# 4. CONFUSION MATRIX 3x3 PER MODEL (PL + NR)
# =========================================================

labels = [0, 1, "NR"]

label_map = {
    0: "Brak klonu",
    1: "Klon",
    "NR": "Brak odpowiedzi"
}

models = df["model"].unique()

fig, axes = plt.subplots(3, 3, figsize=(16, 12))
axes = axes.flatten()

for i, m in enumerate(models):
    sub = df[df["model"] == m]

    cm = pd.crosstab(
        sub["y_true"],
        sub["y_pred"],
        dropna=False
    ).reindex(index=[0, 1], columns=[0, 1, "NR"], fill_value=0)

    ax = axes[i]
    ax.imshow(cm.values, cmap="YlOrRd")

    # --- POWIĘKSZONE: Tytuł podwykresu (nazwa modelu) ---
    ax.set_title(m, fontsize=15, fontweight="bold")

    # --- POWIĘKSZONE: Główne etykiety osi (Predykcja / Rzeczywistość) ---
    ax.set_xlabel("Predykcja", fontsize=13)
    ax.set_ylabel("Rzeczywistość", fontsize=13)
    
    ax.set_xticks([0, 1, 2])
    ax.set_yticks([0, 1])

    # --- POWIĘKSZONE: Etykiety klas (Brak klonu, Klon, Brak odpowiedzi) ---
    ax.set_xticklabels([label_map[x] for x in labels], rotation=30, ha="right", fontsize=12)
    ax.set_yticklabels([label_map[x] for x in [0, 1]], fontsize=12)

    for (r, c), val in np.ndenumerate(cm.values):
        # --- POWIĘKSZONE: Liczby wewnątrz macierzy (z poziomu 13 na 16) ---
        txt = ax.text(
            c, r, str(val),
            ha="center",
            va="center",
            fontsize=16,  
            fontweight="bold",
            color="black"
        )

        txt.set_path_effects([
            path_effects.Stroke(linewidth=3, foreground="white"),
            path_effects.Normal()
        ])

# --- POWIĘKSZONE: Główny tytuł całego wykresu ---
plt.suptitle("Macierze pomyłek (0 / 1 / brak odpowiedzi)", fontsize=20, fontweight="bold")
fig.subplots_adjust(hspace=0.5)
plt.tight_layout(rect=[0.03, 0.03, 1, 0.95])
plt.show()


# =========================================================
# 5. BONUS: NR RATE per model
# =========================================================

plt.figure()
plt.bar(metrics_df["model"], metrics_df["nr_rate"])
plt.title("Odsetek braków odpowiedzi (NR) per model")
plt.ylabel("NR rate")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()


# =========================================================
# 6. BONUS: Precision / Recall / F1 / Accuracy / Specificity
# =========================================================

plt.figure()

x = range(len(metrics_df))

# Użycie formatu "o" usuwa linię i pozostawia same punkty
plt.plot(x, metrics_df["precision"], "o", label="Precision")
plt.plot(x, metrics_df["recall"], "o", label="Recall")
plt.plot(x, metrics_df["f1"], "o", label="F1")
plt.plot(x, metrics_df["accuracy"], "o", label="Accuracy")
plt.plot(x, metrics_df["specificity"], "o", label="Specificity")

plt.xticks(x, metrics_df["model"], rotation=45, ha="right")
plt.title("Precision / Recall / F1 / Accuracy / Specificity per model")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModel

# 1. Ładowanie modelu za pomocą czystego transformers i torch
device = "cuda" if torch.cuda.is_available() else "cpu"
tokenizer = AutoTokenizer.from_pretrained('BAAI/bge-small-en-v1.5')
model = AutoModel.from_pretrained('BAAI/bge-small-en-v1.5').to(device)

def get_embeddings(texts, batch_size=32):
    all_embeddings = []
    for i in range(0, len(texts), batch_size):
        batch_texts = texts[i:i+batch_size]
        encoded_input = tokenizer(batch_texts, padding=True, truncation=True, max_length=512, return_tensors='pt').to(device)
        
        with torch.no_grad():
            model_output = model(**encoded_input)
            # Wyciągamy CLS token i normalizujemy wektor
            sentence_embeddings = model_output[0][:, 0]
            sentence_embeddings = F.normalize(sentence_embeddings, p=2, dim=1)
            all_embeddings.append(sentence_embeddings.cpu().numpy())
            
    return np.vstack(all_embeddings)

print("Generowanie embeddingów...")
responses = descriptions_df["response"].fillna("").tolist()
embeddings = get_embeddings(responses)

# 2. Obliczanie podobieństwa cosinusowego parami (iloczyn skalarny dla znormalizowanych wektorów)
similarities = []
true_clones = []

for i in range(0, len(descriptions_df) - 1, 2):
    emb1 = embeddings[i]
    emb2 = embeddings[i+1]
    
    sim = np.dot(emb1, emb2)
    similarities.append(sim)
    
    is_clone_truth = int(descriptions_df.iloc[i]["is_clone"])
    true_clones.append(is_clone_truth)

similarities = np.array(similarities)
true_clones = np.array(true_clones)

# 3. Ręczne obliczanie metryk (rozszerzone o Accuracy i Specificity)
thresholds = np.linspace(0.0, 1.0, 100)
precisions, recalls, f1_scores = [], [], []
accuracies, specificities = [], []

for t in thresholds:
    predictions = (similarities >= t).astype(int)
    
    # Wyliczanie TP, FP, FN, TN
    tp = np.sum((predictions == 1) & (true_clones == 1))
    fp = np.sum((predictions == 1) & (true_clones == 0))
    fn = np.sum((predictions == 0) & (true_clones == 1))
    tn = np.sum((predictions == 0) & (true_clones == 0))
    
    # Podstawowe metryki
    p = tp / (tp + fp) if (tp + fp) > 0 else 0.0
    r = tp / (tp + fn) if (tp + fn) > 0 else 0.0
    f1 = 2 * (p * r) / (p + r) if (p + r) > 0 else 0.0
    
    # Nowe metryki: Accuracy i Specificity
    acc = (tp + tn) / (tp + tn + fp + fn) if (tp + tn + fp + fn) > 0 else 0.0
    spec = tn / (tn + fp) if (tn + fp) > 0 else 0.0
    
    precisions.append(p)
    recalls.append(r)
    f1_scores.append(f1)
    accuracies.append(acc)
    specificities.append(spec)

# 4. Rysowanie wykresów (2 panele obok siebie)
plt.figure(figsize=(15, 6))

# Wykres 1: F1-Score
plt.subplot(1, 2, 1)
plt.plot(thresholds, f1_scores, label="F1-Score", color="purple", linewidth=2.5)
best_idx = np.argmax(f1_scores)
plt.axvline(thresholds[best_idx], color="red", linestyle="--", 
            label=f"Best Threshold F1: {thresholds[best_idx]:.2f}\nMax F1: {f1_scores[best_idx]:.2f}")
plt.title("Cosine Similarity Threshold vs F1-Score", fontsize=12)
plt.xlabel("Próg (Threshold)", fontsize=10)
plt.ylabel("F1-Score", fontsize=10)
plt.grid(True, linestyle=":", alpha=0.6)
plt.legend(fontsize=10)

# Wykres 2: Wszystkie metryki razem (Precision, Recall, Accuracy, Specificity)
plt.subplot(1, 2, 2)
plt.plot(thresholds, precisions, label="Precision", color="blue", linewidth=2)
plt.plot(thresholds, recalls, label="Recall", color="orange", linewidth=2)
plt.plot(thresholds, accuracies, label="Accuracy", color="green", linewidth=2, linestyle="-.")
plt.plot(thresholds, specificities, label="Specificity (Swoistość)", color="crimson", linewidth=2, linestyle=":")

plt.title("Cosine Similarity Threshold vs All Metrics", fontsize=12)
plt.xlabel("Próg (Threshold)", fontsize=10)
plt.ylabel("Wartość metryki", fontsize=10)
plt.grid(True, linestyle=":", alpha=0.6)
plt.legend(fontsize=10, loc="lower left") # Umieszczone na dole po lewej, żeby nie zasłaniało wykresu przy t -> 1.0

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModel

device = "cuda" if torch.cuda.is_available() else "cpu"

responses = descriptions_df["response"].fillna("").tolist()
true_clones = np.array([int(descriptions_df.iloc[i]["is_clone"])
                        for i in range(0, len(descriptions_df)-1, 2)])


# ==========================================
# METRYKI
# ==========================================
def oblicz_metryki(similarities, true_clones):
    thresholds = np.linspace(0.0, 1.0, 100)
    p_list, r_list, f1_list, acc_list, spec_list = [], [], [], [], []

    for t in thresholds:
        preds = (similarities >= t).astype(int)

        tp = np.sum((preds == 1) & (true_clones == 1))
        fp = np.sum((preds == 1) & (true_clones == 0))
        fn = np.sum((preds == 0) & (true_clones == 1))
        tn = np.sum((preds == 0) & (true_clones == 0))

        p = tp / (tp + fp) if (tp + fp) else 0
        r = tp / (tp + fn) if (tp + fn) else 0
        f1 = 2 * p * r / (p + r) if (p + r) else 0
        acc = (tp + tn) / (tp + tn + fp + fn)
        spec = tn / (tn + fp) if (tn + fp) else 0

        p_list.append(p)
        r_list.append(r)
        f1_list.append(f1)
        acc_list.append(acc)
        spec_list.append(spec)

    return thresholds, p_list, r_list, f1_list, acc_list, spec_list


# ==========================================
# BGE MODEL
# ==========================================
print("Przetwarzanie BGE-small...")

tokenizer_bge = AutoTokenizer.from_pretrained("BAAI/bge-small-en-v1.5")
model_bge = AutoModel.from_pretrained("BAAI/bge-small-en-v1.5").to(device)
model_bge.eval()


def get_embeddings_bge(texts, batch_size=32):
    embs_all = []

    for i in range(0, len(texts), batch_size):
        batch = texts[i:i+batch_size]

        encoded = tokenizer_bge(
            batch,
            padding=True,
            truncation=True,
            max_length=512,
            return_tensors="pt"
        ).to(device)

        with torch.no_grad():
            out = model_bge(**encoded)

            # CLS + normalizacja
            embs = F.normalize(out.last_hidden_state[:, 0], p=2, dim=1)

        embs_all.append(embs.cpu().numpy())

    return np.vstack(embs_all)


embs_bge = get_embeddings_bge(responses)


# ==========================================
# SIMILARITY (pary)
# ==========================================
sims_bge = []

for i in range(0, len(responses)-1, 2):
    sims_bge.append(np.dot(embs_bge[i], embs_bge[i+1]))

sims_bge = np.array(sims_bge)


# ==========================================
# METRYKI
# ==========================================
t_bge, p_bge, r_bge, f1_bge, acc_bge, spec_bge = oblicz_metryki(sims_bge, true_clones)


# ==========================================
# WYKRESY (IDENTYCZNY STYL JAK CODEBERT)
# ==========================================
plt.figure(figsize=(16, 12))


# --- PANEL 1: F1 ---
plt.subplot(2, 2, 1)
plt.plot(t_bge, f1_bge, label="F1-Score", color="purple", linewidth=2.5)

best_idx = np.argmax(f1_bge)
plt.axvline(
    t_bge[best_idx],
    color="red",
    linestyle="--",
    label=f"Best Thresh: {t_bge[best_idx]:.2f}\nMax F1: {f1_bge[best_idx]:.2f}"
)

plt.title("BGE-small: Threshold vs F1-Score")
plt.xlabel("Próg")
plt.ylabel("F1")
plt.grid(True, linestyle=":", alpha=0.6)
plt.legend()


# --- PANEL 2: METRYKI ---
plt.subplot(2, 2, 2)
plt.plot(t_bge, p_bge, label="Precision", color="blue", linewidth=2)
plt.plot(t_bge, r_bge, label="Recall", color="orange", linewidth=2)
plt.plot(t_bge, acc_bge, label="Accuracy", color="green", linewidth=2, linestyle="-.")
plt.plot(t_bge, spec_bge, label="Specificity", color="crimson", linewidth=2, linestyle=":")

plt.title("BGE-small: Threshold vs All Metrics")
plt.xlabel("Próg")
plt.ylabel("Wartość")
plt.grid(True, linestyle=":", alpha=0.6)
plt.legend(loc="lower left")


# --- PANEL 3: KRZYWA F1 ---
plt.subplot(2, 2, 3)
plt.plot(t_bge, f1_bge, label="F1-Score", color="purple", linewidth=2.5)
plt.title("BGE-small: F1 Curve")
plt.xlabel("Próg")
plt.ylabel("F1")
plt.grid(True, linestyle=":", alpha=0.6)
plt.legend()


# --- PANEL 4: ROZKŁAD SIMILARITY ---
plt.subplot(2, 2, 4)

plt.hist(sims_bge[true_clones == 1], bins=30, alpha=0.6, label="Clones (1)", color="green")
plt.hist(sims_bge[true_clones == 0], bins=30, alpha=0.6, label="Non-clones (0)", color="red")

plt.title("BGE-small: Similarity Distribution")
plt.xlabel("Cosine similarity")
plt.ylabel("Count")
plt.grid(True, linestyle=":", alpha=0.6)
plt.legend()


plt.tight_layout()
plt.show()
```

```python
import numpy as np
import torch
import torch.nn.functional as F
from transformers import AutoTokenizer, AutoModel
from tqdm import tqdm

device = "cuda" if torch.cuda.is_available() else "cpu"

# 1. Model
tokenizer = AutoTokenizer.from_pretrained("microsoft/codebert-base")
model = AutoModel.from_pretrained("microsoft/codebert-base").to(device)
model.eval()


# 2. Dane (pary)
prompty_A = descriptions_df["prompt"].iloc[0::2].fillna("").tolist()
prompty_B = descriptions_df["prompt"].iloc[1::2].fillna("").tolist()
true_clones = descriptions_df["is_clone"].iloc[0::2].astype(int).to_numpy()


# 3. Funkcja do embeddingu
def encode(texts, batch_size=16):
    embeddings = []

    for i in tqdm(range(0, len(texts), batch_size)):
        batch = texts[i:i + batch_size]

        inputs = tokenizer(
            batch,
            padding=True,
            truncation=True,
            max_length=512,
            return_tensors="pt"
        ).to(device)

        with torch.no_grad():
            outputs = model(**inputs)

            # pooling: średnia z tokenów (lepsze niż CLS w bi-encoderze)
            attention_mask = inputs["attention_mask"]
            token_embeddings = outputs.last_hidden_state

            mask = attention_mask.unsqueeze(-1).expand(token_embeddings.size()).float()
            summed = torch.sum(token_embeddings * mask, dim=1)
            counts = torch.clamp(mask.sum(dim=1), min=1e-9)

            mean_embeddings = summed / counts

            embeddings.append(mean_embeddings.cpu())

    return torch.cat(embeddings, dim=0)


# 4. Embeddingi osobno
print("Kodowanie A...")
emb_A = encode(prompty_A)

print("Kodowanie B...")
emb_B = encode(prompty_B)


# 5. Cosine similarity
similarities = F.cosine_similarity(emb_A, emb_B).numpy()


# 6. Normalizacja (opcjonalnie)
similarities = (similarities - similarities.min()) / (similarities.max() - similarities.min() + 1e-9)


# 7. Threshold evaluation
thresholds = np.linspace(0, 1, 100)

precisions, recalls, f1_scores, accuracies, specificities = [], [], [], [], []

for t in thresholds:
    preds = (similarities >= t).astype(int)

    tp = np.sum((preds == 1) & (true_clones == 1))
    fp = np.sum((preds == 1) & (true_clones == 0))
    fn = np.sum((preds == 0) & (true_clones == 1))
    tn = np.sum((preds == 0) & (true_clones == 0))

    p = tp / (tp + fp) if (tp + fp) else 0
    r = tp / (tp + fn) if (tp + fn) else 0
    f1 = 2 * p * r / (p + r) if (p + r) else 0
    acc = (tp + tn) / (tp + tn + fp + fn)
    spec = tn / (tn + fp) if (tn + fp) else 0

    precisions.append(p)
    recalls.append(r)
    f1_scores.append(f1)
    accuracies.append(acc)
    specificities.append(spec)
```
